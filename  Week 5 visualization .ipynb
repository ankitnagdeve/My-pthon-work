```python
def  samplechart():
    x= np.arange(10)
    y= x**2
    plt.plot(x,y , color = 'red')
    plt.title('sample chart')
    plt.xlabel('series')
    plt.ylabel('its square')
    plt.show();
samplechart()
```

```python
def sine_wave():
    x = np.linspace(-np.pi,(np.pi)+1)
    y= np.sin(x)
    plt.plot(x,y)
    plt.show(x,y)
```

```python
for i, speed in enumerate([9, 11, 15, 13]):
    
    # Set up the plot
    ax = plt.subplot(2, 2, i + 1)
    
    # Draw the plot
    ax.hist(wea['Wind Spd (km/h)'], bins = int(180/speed),
             color = 'blue', edgecolor = 'black')
    
    # Title and labels
    ax.set_title('speed = %d' % speed, size = 30)
    ax.set_xlabel('speed', size = 12)
    ax.set_ylabel('a', size= 12)

plt.tight_layout()
```

```python
names = ['Mainly Clear .', 'Mostly Cloudy', 'Cloudy',
         'Clear', 'Snow']

# Iterate through the five airlines
for names in names:
    # Subset to the airline
    subset = wea[wea['Weather'] == names]
    
    # Draw the density plot
    sns.distplot(subset['Rel Hum (%)'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = names)
    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'Temp of weather')
plt.title('Density Plot with top weather')
plt.xlabel('Density ')
plt.ylabel('temp')
```
